```python
def lshow(img, vmax=None):
    """
    Lambert's version of imshow.
    """
    f,ax = plt.subplots()
    cmap = None
    if (len(img.shape) == 2) or (img.shape[2] == 1):
        cmap = 'viridis'
    him = ax.imshow(img, cmap=cmap, vmax=vmax)
    if img.dtype != bool:
        f.colorbar(him, ax=ax)
    return ax
```

```python
ax = lshow(img2)
```

```python
def clear_graphics(ax):
    if ax.lines:
        [lin.remove() for lin in reversed(ax.lines)]
    # Can do similar removals with ax.patches, and with objects returned by ax.get_children() such
    # as instances of mpl.text.Annotation and mpl.collections.PathCollection
```

```python
def random_lines(ax, num_lines, bound):
    for i in range(num_lines):
        x1 = random.randint(0,bound)
        x2 = random.randint(0,bound)
        y1 = random.randint(0,bound)
        y2 = random.randint(0,bound)
        mlin = mpl.lines.Line2D((x1, x2), (y1, y2), marker='x')
        ax.add_line(mlin)
```

```python
clear_graphics(ax)
random_lines(ax, 5, 1000)
```
